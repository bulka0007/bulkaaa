```python
arr = np.array([1, 2, 4, 5])

total = arr.sum()
left_sums = np.cumsum(arr) - arr          # сумма элементов до i
right_sums = total - left_sums - arr     # сумма элементов после i
diffs = np.abs(left_sums - right_sums)   # модуль разности
best_idx = np.argmin(diffs)              # индекс с минимальной разницей

print(best_idx)   # 2
print(arr[best_idx])  # 4
```
